```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///breweries.sqlite')
conn=engine.connect()
pd.read_csv('breweries.csv', index_col='index').to_sql('breweries', conn, if_exists='replace')
```

```python
conn.execute('pragma table_info(breweries);').fetchall()
```

```python
sql_rename_query='ALTER TABLE breweries RENAME TO breweries_old; '
sql_create_query='CREATE TABLE breweries (\
                  id bigint primary key,\
                  name text,\
                  website text,\
                  address text,\
                  city text,\
                  state text\
                  ); '
sql_insert_query='INSERT INTO breweries SELECT * FROM breweries_old; '
sql_drop_query='DROP TABLE breweries_old'
```

```python
conn.execute(sql_rename_query)
conn.execute(sql_create_query)
conn.execute(sql_insert_query)
conn.execute(sql_drop_query)
```

```python
conn.execute('select * from breweries limit 10').fetchall()
```

```python
from sqlalchemy.orm import Session
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
```

```python
Base=automap_base()
Base.prepare(engine, reflect=True)
Brewery=Base.classes.breweries

session=Session(engine)
```

```python
results=session.query(Brewery).all()
```
